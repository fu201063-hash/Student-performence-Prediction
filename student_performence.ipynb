```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('Student_Performance - Student_Performance.csv')
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
df.info()
```

ENCODING

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'Yes':1,'No':0})
```

```python
df['Extracurricular Activities']
```

DIVIDE DATASET INTO INPUT AND OUTPUT

```python
input = df.drop('Performance Index',axis=1)
```

```python
input
```

```python
output = df['Performance Index']
output
```

SCALLING THE DATA

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = train_test_split(input,output,test_size = 0.25,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
scaled = StandardScaler()
```

```python
scaled_train = scaled.fit_transform(x_train)
scaled_test = scaled.transform(x_test)
```

```python
scaled_test
```

```python
scaled_train
```

MODELING

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(scaled_train,y_train)
```

```python
y_train_scaled = model.predict(scaled_train)
y_train_scaled
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_test_scaled = model.predict(scaled_test)
y_test_scaled
```

```python
from sklearn.metrics import mean_squared_error
```

```python
train_mse = mean_squared_error(y_train,y_train_scaled)
train_mse
```

```python
rmse = np.sqrt(train_mse)
rmse
```

```python
test_mse = mean_squared_error(y_test,y_test_scaled)
test_mse
```

```python
rmse = np.sqrt(test_mse)
rmse
```

**R2 SCORE**

```python
model.score(scaled_train,y_train)
```
